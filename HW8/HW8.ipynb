A Classification Decision Tree addresses categorical prediction problems.
The goal of Classification Decision Tree is to assign data points to one of a finite set of classes or categories based on input features

It works by splitting the dataset into branches based on feature values, ultimately leading to a decision at a "leaf" (the end node of the tree) that represents a class. And the output (dependent variable) is categorical.



For the difference of making predictions between Classification Decision Tree and Multiple Linear Regression is as follow:


| **Aspect**                  | **Classification Decision Tree**                                             | **Multiple Linear Regression**                                              |
|-----------------------------|-----------------------------------------------------------------------------|-----------------------------------------------------------------------------|
| **Prediction Type**         | Makes **categorical predictions** (e.g., Class A, B, or C).                  | Makes **continuous predictions** (e.g., predicting housing prices).         |
| **Output Variable**         | Target variable is **categorical**.                                          | Target variable is **numerical/continuous**.                                |
| **How Predictions are Made**| Splits the data into **decision nodes** based on feature thresholds, leading to class labels. | Finds the **best-fit line** (or hyperplane) to minimize the error (e.g., least squares). |
| **Mathematical Basis**      | Uses **information gain**, **Gini impurity**, or **entropy** to decide splits. | Uses **linear relationships** between features and the target variable.     |
| **Data Interpretation**     | Decision rules (e.g., "If Age > 30 and Income < $40K, then Class B").         | Coefficients show how much the target changes with one-unit change in features. |
| **Handle Nonlinear Data?**  | Works well with **nonlinear relationships**.                                 | Assumes a **linear relationship** between features and target.              |
| **Flexibility**             | Can handle both categorical and numerical features.                          | Primarily handles numerical features (requires encoding for categoricals).  |

## *1. **Accuracy*** 

measures the proportion of true results (both true positives and true negatives) in the population. 

Or simply saying, it measures how often the model gets it right. High accuracy indicates that the model is performing well overall, but it may not always reflect its true usefulness, especially in imbalanced datasets.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$

*Appropriate when the dataset is balanced (similar numbers of positive and negative cases) and all types of errors are equally important, then use this.*

**The dataset is balanced** means both positive and negative classes are present in roughly equal proportions.

**All types of errors (false positives and false negatives) have similar costs(equally important)**: Misclassifications in either direction are equally undesirable.

Example: Quality Control in Manufacturing

A decision tree predicts whether a product is Defective or Non-Defective. Accuracy is suitable here since defective products (false negatives) and wrongly flagged products (false positives) carry similar costs.

If 10% of products are defective and 90% are not, accuracy will perform well as a metric if the costs of false positives (falsely flagging a good product as defective) and false negatives (missing a defective product) are similar.

### **Real-World Example: Predicting Loan Default**
Imagine a bank using a decision tree to predict whether a loan applicant will **default** or not. Suppose:
- 10,000 total applicants.
- 9,500 do not default, 500 do default.

#### **Model Performance**:
- The model predicts:
  - 9,300 as "No Default" (9,200 correct, 100 incorrect).
  - 700 as "Default" (300 correct, 400 incorrect).

**Confusion Matrix**:

|                | **Predicted: No Default** | **Predicted: Default** |
|----------------|---------------------------|-------------------------|
| **Actual: No Default** | 9,200                  | 300                     |
| **Actual: Default**    | 100                    | 400                     |

#### **Accuracy Calculation**:
$\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{Total Predictions}} = \frac{400 + 9,200}{10,000} = 96\%$

At first glance, this model seems excellent (96% accuracy), but:
- **Sensitivity** = $\frac{\text{TP}}{\text{TP} + \text{FN}} = \frac{400}{400 + 100} = 80\%$: It misses 20% of defaulters.
- **Specificity** = $\frac{\text{TN}}{\text{TN} + \text{FP}} = \frac{9,200}{9,200 + 300} = 96.8\%$: Good at identifying non-defaulters.


## *2. **Sensitivity***

measures the proportion of actual positives that are correctly identified.

Sensitivity, also called Recall or the True Positive Rate (TPR), measures the ability of a classification model to correctly identify positive instances in the dataset. We can say it answers: Of all the actual positives in the dataset, how many did the model correctly predict as positive?

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$



*Sensitivity is particularly important in scenarios where missing positive cases (false negatives) carries a high cost or risk.*

- Example 1: Medical Diagnosis: Predicting whether a patient has Cancer (positive class).

Missing a true cancer case (FN) can delay treatment, so the model must identify as many positives as possible, even at the expense of some false alarms.

- Example 2: Fraud Detection: Classifying credit card transactions as Fraudulent or Legitimate.

Missing fraudulent cases (FN) can lead to significant financial loss, so high sensitivity is crucial.


### **Example: Sensitivity Calculation**

#### **Scenario**:
A Decision Tree model predicts whether a patient has a disease:
- **Actual Cases**:
  - 100 patients have the disease.
  - 900 patients do not have the disease.
- **Model Predictions**:
  - Correctly predicts 90 patients as having the disease (**TP = 90**).
  - Misses 10 patients who actually have the disease (**FN = 10**).

#### **Confusion Matrix**:

|                       | **Predicted: Disease** | **Predicted: No Disease** |
|-----------------------|------------------------|---------------------------|
| **Actual: Disease**   | **TP = 90**            | **FN = 10**               |
| **Actual: No Disease**| FP = 50                | TN = 850                  |

#### **Sensitivity Formula**:
$\text{Sensitivity} = \frac{\text{TP}}{\text{TP} + \text{FN}} = \frac{90}{90 + 10} = 0.9 \, (90\%)$

**Interpretation**:  
- The model correctly identifies 90% of patients with the disease.
- However, it misses 10% of patients who actually have the disease (FN).

### **In summary:**

Sensitivity focuses on minimizing false negatives.

It is critical in high-stakes applications like medical diagnosis, fraud detection, or safety inspections.


## *3. **Specificity***

measures the proportion of actual negatives that are correctly identified.

Specificity also known as the True Negative Rate (TNR), measures the ability of a classification model to correctly identify negative instances in the dataset. Specificity answers: Of all the actual negatives in the dataset, how many did the model correctly predict as negative?

$$\text{Specificity} = \frac{TN}{TN + FP}$$



*Specificity is critical in scenarios where false positives (wrongly classifying a negative instance as positive) carry significant consequences or costs.*

1. Spam Email Classification

Scenario: Classify emails as "Spam" or "Not Spam."

Why Specificity Matters:

    - A false positive means labeling a legitimate email as spam, causing the user to miss important messages.

    - High specificity ensures legitimate emails (negatives) are correctly left in the inbox.

2. Criminal Investigation

Scenario: Predict whether a suspect is "Guilty" or "Innocent."

Why Specificity Matters:
    
    - A false positive means wrongly classifying an innocent person as guilty, leading to severe ethical and legal consequences.

    - High specificity minimizes wrongful accusations, ensuring innocent individuals are not falsely flagged.

3. Safety Testing

Scenario: Predict whether a car part is "Faulty" or "Non-Faulty."

Why Specificity Matters:

    - A false positive means discarding a functional part, leading to increased costs and waste.

    - High specificity ensures that functional parts are not mistakenly rejected.

### **Specificity in Action**

#### **Scenario**: Fraudulent Transactions
A bank uses a decision tree to predict whether a transaction is **Fraudulent (Positive)** or **Legitimate (Negative)**.  
- **Dataset**: 10,000 transactions.
  - 500 are fraudulent, 9,500 are legitimate.

#### **Model Predictions**:
- Correctly identifies 9,300 legitimate transactions (**TN = 9,300**).
- Incorrectly labels 200 legitimate transactions as fraudulent (**FP = 200**).

#### **Confusion Matrix**:

|                       | **Predicted: Fraudulent** | **Predicted: Legitimate** |
|-----------------------|--------------------------|---------------------------|
| **Actual: Fraudulent**| TP = 400                 | FN = 100                  |
| **Actual: Legitimate**| **FP = 200**             | **TN = 9,300**            |

#### **Specificity Formula**:
$\text{Specificity} = \frac{\text{TN}}{\text{TN} + \text{FP}} = \frac{9,300}{9,300 + 200} = 0.979 \, (97.9\%)$

**Interpretation**:
- The model correctly identifies 97.9% of legitimate transactions.
- Only 2.1% of legitimate transactions are wrongly flagged as fraudulent.


## *4. **Precision*** 

measures the proportion of positive identifications that were actually correct.

Precision, also known as the Positive Predictive Value (PPV), measures how many of the predicted positive cases are actually true positives. Precision answers: When the model predicts something as positive, how often is it correct?

$$\text{Precision} = \frac{TP}{TP + FP}$$

*Precision is crucial when false positives (FP) are costly or problematic.*

1. Email Spam Detection

Scenario: Classify emails as "Spam" or "Not Spam."

Why Precision Matters:

    - A false positive (classifying a legitimate email as spam) can result in important emails being missed by the user.

    - High precision ensures that emails flagged as spam are almost always actual spam, minimizing disruption for the user.

2. Fraud Detection

Scenario: Identify fraudulent transactions.

Why Precision Matters:

    - A false positive means flagging a legitimate transaction as fraudulent, causing inconvenience to customers (e.g., credit card blocks).

    - High precision ensures that flagged transactions are almost always fraudulent, minimizing customer frustration.

3. Online Advertising

Scenario: Predict whether a user will click on an ad.

Why Precision Matters:

    - A false positive (predicting a user will click but they don’t) wastes advertising dollars on uninterested users.

    - High precision ensures that targeted ads are likely to reach users who will engage with them.

4. Candidate Screening for Jobs

Scenario: Predict whether a candidate is a good fit for a job.

Why Precision Matters:

    - A false positive means inviting unqualified candidates for interviews, wasting time and resources.

    - High precision ensures that selected candidates are highly likely to meet the job requirements.

### **Example: Precision Calculation**

#### **Scenario**:
A decision tree predicts whether an email is **Spam (Positive)** or **Not Spam (Negative)**.
- **Actual Cases**:
  - 500 emails are actually spam.
  - 9,500 emails are not spam.
- **Model Predictions**:
  - Predicts 600 emails as spam.
  - Of these, 400 are correctly identified as spam (**TP = 400**), but 200 are false positives (**FP = 200**).

#### **Confusion Matrix**:

|                       | **Predicted: Spam** | **Predicted: Not Spam** |
|-----------------------|---------------------|--------------------------|
| **Actual: Spam**      | **TP = 400**        | FN = 100                |
| **Actual: Not Spam**  | **FP = 200**        | TN = 9,300              |

#### **Precision Formula**:
$\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}} = \frac{400}{400 + 200} = 0.666 \, (66.6\%)$

**Interpretation**:
- Of all the emails predicted as spam, only 66.6% were actual spam.
- The remaining 33.4% were false positives (legitimate emails mistakenly flagged as spam).

At each node, a binary split is made based on one feature and a threshold. Each leaf represents the final decision, and the class with the highest proportion of training examples in that leaf is chosen as the prediction.

## 9. Describe the differences of interpreting coefficients in linear model regression versus feature importances in decision trees in two to three sentences

In linear regression, coefficients directly represent the magnitude and direction of influence that each predictor variable has on the predicted outcome, assuming all other variables are held constant. This makes interpreting the contribution of each variable straightforward and transparent. 

In decision trees, feature importances summarize how much each variable contributes to improving the predictive power of the model by reducing impurity at each split. However, this contribution is aggregated across the entire tree and includes interactions between variables, making the interpretation less direct and harder to isolate compared to linear regression.
