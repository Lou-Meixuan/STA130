### 3. The second "Pre-lecture" video (above) explains that we "imagine a world where the null hypothesis is true" when calculating a p-value? Explain why this is in your own words in a way that makes the most sense to you.

When calculating a p-value, we start by assuming that the null hypothesis is true. We can use the sampling distribution of the test statistic to create a model of what we would expect to see if the null hypothesis were true.

The sampling distribution represents the range of possible outcomes for our test statistic if we repeatedly took samples from the population under the null hypothesis. This distribution helps us determine how extreme our observed result is. 

For instance, if we find that our observed test statistic falls far into the tail of this distribution which means that it has a very low probability of occurring, it suggests that we reject the null hypothesis. Conversely, if our observed statistic is close to the center of the distribution, it implies that we fail to reject the null hypothesis. 

Thus, imagining a world where the null hypothesis is true help us to define whether our observed data is unusual, guiding our decision about whether to reject or fail to reject the null hypothesis based on the calculated p-value.

### 5. Gntrkn (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a **p-value** using a "50/50 coin-flipping" model for the assumption of the **null hypothesis** $H_0$ that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against $H_0$. <br>

$H_0$: the population of humans don't have left or right head tilt tendencies when kissing.

```python
import numpy as np

# Parameters
n_trials = 10000  # Number of simulations
n_flips = 124     # Total number of couples
observed_heads = 80  # Number of couples that tilted their heads to the right

# Simulate coin flips (0 for left tilt, 1 for right tilt)
simulated_heads = np.random.binomial(n_flips, 0.5, n_trials)

# Calculate the p-value
p_value = np.mean(simulated_heads >= observed_heads)

# Print the result
print(f"P-value: {p_value:.4f}")
```


|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

Since the p-value we get is smaller then 0.001, which means we have very strong evidence against the null hypothesisthat. So that we reject the null hypothesis. 
This means that the population of humans have left or right head tilt tendencies when kissing.

### 8. Complete the following assignment.

### Problem Introduction

#### overview
We are replicating the famous Fisher and Bristol tea-tasting scenario, but with a sample of 80 students from STA130 rather than single individual. The objective is to determine if students can identify whether milk or tea was poured first, or if their correct responses could be attributed to random guessing.

#### Statements of the Null Hypothesis and Alternative Hypothesis
"p" represents the true proportion of students in the population who can correctly identify the pouring order by chance alone.


- Null Hypothesis ($H_0$): The students are simply guessing, with a 50% probability of identifying whether milk or tea was poured first.

$$H_0: \text{ p = 0.5}$$

- Alternative Hypothesis ($H_A$): The students are not guessing and have a greater than 50% chance of correctly identifying the order.

$$H_0: \text{ p > 0.5}$$




### Quantitative Analysis

We would now conduct the analysis by calculating the test statistic and determining whether the observed result (49 out of 80 students correct) provides evidence against the null hypothesis.

Observed Test Statistic: $\bar x$ = 49/80 = 0.6125

Null Hypothesis Parameter: $\mu_0$ = 0.5

Then we will do the hypothesis testing. Simulate a p-value using a "50/50 coin-flipping" model for the assumption of the null hypothesis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
n = 80                # sample size (number of students)
X = 49                # number of correct identifications
p_0 = 0.5             # null hypothesis proportion
num_simulations = 10000  # number of simulated trials

# Step 1: Simulate coin flips
simulated_successes = np.random.binomial(n, p_0, num_simulations)

# Step 2: Calculate the proportion of successes in the simulations
simulated_proportions = simulated_successes / n

# Step 3: Calculate the observed proportion
observed_proportion = X / n

# Step 4: Calculate the p-value
p_value = np.mean(simulated_proportions >= observed_proportion)

# Visualization
plt.figure(figsize=(10, 6))
plt.hist(simulated_proportions, bins=30, alpha=0.7, color='blue', density=True)
plt.axvline(observed_proportion, color='red', linestyle='--', label=f'Observed Proportion: {observed_proportion:.2f}')
plt.title('Distribution of Simulated Proportions of Success')
plt.xlabel('Proportion of Success')
plt.ylabel('Density')
plt.legend()
plt.grid()

# Show the plot
plt.show()

# Output results
print(f"Observed Proportion: {observed_proportion:.4f}")
print(f"Simulated P-Value: {p_value:.4f}")

# Conclusion based on p-value
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The students are not just guessing.")
else:
    print("Fail to reject the null hypothesis: No evidence that the students are not just guessing.")
```

### Findings and Discussion

The p-value we get is smaller than 0.05

|-|-|

|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|    

Based on the analysis and the table, we can reject the null hypothesis. Which means that the students are not guessing and have a greater than 50% chance of correctly identifying the order.
